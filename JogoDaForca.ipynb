# Jogo da Forca

```python
import unicodedata
import pandas as pd
import random

HANGMANPICS = [

    '''
    +---+
    |   |
        |
        |
        |
        |
   =========''', '''
    +---+
    |   |
    o   |
        |
        |
        |
   =========''', '''
    +---+
    |   |
    o   |
    |   |
        |
        |
   =========''', '''
    +---+
    |   |
    o   |
   /|   |
        |
        |
   =========''', '''
    +---+
    |   |
    o   |
   /|\  |
        |
        |
   =========''', '''
    +---+
    |   |
    o   |
   /|\  |
   /    |
        |
   ========''', '''
    +---+
    |   |
    o   |
   /|\  |
   / \  |
        |
   ========'''
]

# Copiar a partir daqui

# Remove acentuação e passa para minuscula


def strip_accents(text):

    try:
        text = unicode(text, 'utf-8')
    except NameError:  # unicode is a default on python 3
        pass

    text = unicodedata.normalize('NFD', text)\
        .encode('ascii', 'ignore')\
        .decode("utf-8")

    return str(text).lower()


# Base de dados de palavras
df = pd.read_csv('forca_palavras.csv', header=None, encoding='latin-1')

# Palavra escolhida
randomico = random.randint(0, len(df) - 1)
palavra_escolhida = strip_accents(df.iloc[randomico][0])
dica = df.iloc[randomico][1]

# Palavra a ser revelada
palavra_atualizada = [' _ ' if i not in [' ', ',', '.',
                                         '-', '_'] else ' '+i+' ' for i in list(palavra_escolhida)]
print('DICA: ' + str(dica))
print(''.join(palavra_atualizada))

# Erro negativo para mostrar todas as imagens da forca
error = -1
rodadas = 0

while error < 6:
    escolha = strip_accents(input('Chute uma letra ou palavra: '))
    rodadas += 1

    # Soletrando
    if len(escolha) == 1:
        encontrado = 0
        for i in range(len(list(palavra_escolhida))):
            if list(palavra_escolhida)[i] == escolha:
                palavra_atualizada[i] = escolha
                encontrado = 1
        # Erro
        if encontrado == 0:
            error += 1
            print(error)
            print(HANGMANPICS[error])

    # Chutando
    if len(escolha) > 1:
        if escolha == palavra_escolhida:
            print(palavra_escolhida)
            print('ACERTOU!')
            print('FIM DE JOGO')
            print('Total de rodadas: ' + str(rodadas))
            break
        else:
            error += 1
            print(error)
            print(HANGMANPICS[error])

    resultante = ''.join(palavra_atualizada)
    print(resultante)

    if palavra_escolhida == resultante:
        print('ACERTOU!')
        print('FIM DE JOGO')
        print('Total de rodadas: ' + str(rodadas))
        break

if error >= 6:
    print('VOCÊ PERDEU')
    print('Palavra certa: ' + str(palavra_escolhida))
```
